### Visualization

```python
import matplotlib.pyplot as plt # Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt.
import numpy as np
```

```python
x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.bar(x,y)
plt.show()
```

```python
### The bar() function takes arguments that describes the layout of the bars.

### The categories and their values represented by the first and second argument as arrays.
x = ["Akash", "Bablu"]
y = [7, 6]
plt.bar(x, y)
```

```python
### If you want the bars to be displayed horizontally instead of vertically, use the barh() function.
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.barh(x, y)
plt.show()
```

```python
### The bar() and barh() takes the keyword argument color to set the color of the bars
x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.bar(x, y, color = "pink")
plt.show()
```

```python
x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.bar(x, y, color = "blue")
plt.show()
```

```python
x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.bar(x, y, color = "#4CAF50")
plt.show()
```

```python
x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.bar(x, y, width = 0.7)
plt.show()
```

```python
## The barh() takes the keyword argument height to set the height of the bars
x = np.array(["Akash", "Bablu", "Chandan", "Deepak"])
y = np.array([4,7,2,12])

plt.barh(x, y, height = 0.4)
plt.show()
```

```python
x = np.random.normal(200, 20, 450)
plt.hist(x)
plt.show() 
```

```python
y = np.array([45, 15, 20, 20])

plt.pie(y)
plt.show()
```

```python
### Add labels to the pie chart with the label parameter.

### The label parameter must be an array with one label for each wedge
y = np.array([45, 15, 20, 20])
mylabels = ["Akash", "Bablu", "Chandan", "Deepak"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
### As mentioned the default start angle is at the x-axis, but you can change the start angle 
### by specifying a startangle parameter.

### The startangle parameter is defined with an angle in degrees, default angle is 0:
y = np.array([45, 15, 20, 20])
mylabels = ["Akash", "Bablu", "Chandan", "Deepak"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
y = np.array([45, 15, 20, 20])
mylabels = ["Akash", "Bablu", "Chandan", "Deepak"]
myexplode = [0.2, 0.8, 0.1, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([45, 15, 20, 20])
mylabels = ["Akash", "Bablu", "Chandan", "Deepak"]
mycolors = ["black", "yellow", "blue", "red"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
y = np.array([45, 15, 20, 20])
mylabels = ["Akash", "Bablu", "Chandan", "Deepak"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```
