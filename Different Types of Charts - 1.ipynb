```python
import matplotlib.pyplot as plt # Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt.
import numpy as np
```

```python
### Draw a line in a diagram from position (0,20) to position (10,200):
xpoints = np.array([0, 20])
ypoints = np.array([10, 200])

plt.plot(xpoints, ypoints)
plt.show()
```

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
xpoints = np.array([1, 10])
ypoints = np.array([9,20])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([6,7])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints, marker = '*')
```

```python
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, 'o:g')
plt.show()
```

```python
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, marker = 'o', ms = 10)
plt.show()
```

```python
### You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, marker = 'o', ms = 10, mec = 'r')
plt.show()
```

```python
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, marker = 'o', ms = 10,mfc='b')
plt.show()
```

```python
### Use both the mec and mfc arguments to color of the entire marker
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, marker = 'o', ms = 10, mec = 'k', mfc = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 10, mec = '#4CAF50', mfc = '#4CAF50')
```

```python
plt.plot(ypoints, marker = 'o', ms = 10, mec = 'hotpink', mfc = 'hotpink')
```

```python
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
```

The line style can be written in a shorter syntax:

linestyle can be written as ls.

dotted can be written as :.

dashed can be written as --.

```python
plt.plot(ypoints, ls = ':')
```

```python
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

```python
plt.plot(ypoints, c = '#4CAF50')
```

```python
plt.plot(ypoints, c = 'hotpink')
```

```python
### You can use the keyword argument linewidth or the shorter lw to change the width of the line.

### The value is a floating number, in points
ypoints = np.array([2, 7, 4, 10])

plt.plot(ypoints, linewidth = '10')
plt.show()
```

```python
y1 = np.array([2, 7, 4, 10])
y2 = np.array([4, 2, 5, 12])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

(In the examples above we only specified the points on the y-axis, meaning that the points on the x-axis got the the default values (0, 1, 2, 3).)

The x- and y- values come in pairs

```python
x1 = np.array([1, 3, 6, 9])
y1 = np.array([2, 7, 1, 15])
x2 = np.array([7, 3, 9, 11])
y2 = np.array([5, 8, 1, 18])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
x = np.array([60, 55, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burn")

plt.show()
```

```python
### You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.
x = np.array([60, 55, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

font1 = {'family':'serif','color':'red','size':25}
font2 = {'family':'serif','color':'darkred','size':20}

plt.title("Sports Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burn", fontdict = font2)

plt.plot(x, y)
plt.show()
```

You can use the loc parameter in title() to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

With Pyplot, you can use the grid() function to add grid lines to the plot

```python
x = np.array([60, 55, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])
plt.title("Sports Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burn")

plt.plot(x, y)
plt.show()
```

```python
x = np.array([60, 55, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

plt.title("Sports Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burn")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
### You can use the axis parameter in the grid() function to specify which grid lines to display.

### Legal values are: 'x', 'y', and 'both'. Default value is 'both'
x = np.array([55, 60, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
### Display only grid lines for the y-axis
x = np.array([55, 60, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

```python
### Set Line Properties for the Grid
### You can also set the line properties of the grid, like this: grid
### (color = 'color', linestyle = 'linestyle', linewidth = number).
x = np.array([55, 60, 70, 85, 105, 115, 120, 125, 130, 155])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
#plot 1:
x = np.array([4, 5, 6, 7])
y = np.array([2, 7, 2, 15])

plt.subplot(2, 4, 2)
plt.plot(x,y)

#plot 2:
x = np.array([4, 5, 6, 7])
y = np.array([2, 7, 2, 15])

plt.subplot(2, 4, 4)
plt.plot(x,y)

plt.show()
```

### The subplots() function takes three arguments that describes the layout of the figure.
### The layout is organized in rows and columns, which are represented by the first and second argument.
### The third argument represents the index of the current plot.
plt.subplot(2, 4, 2)

### the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(2, 4, 4)

### the figure has 1 row, 2 columns, and this plot is the second plot.

```python
#plot 1:
x = np.array([0, 2, 4, 6])
y = np.array([3, 8, 1, 12])

plt.subplot(3, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 2, 4, 6])
y = np.array([20, 30, 40, 50])

plt.subplot(3, 1, 3)
plt.plot(x,y)

plt.show()
```
